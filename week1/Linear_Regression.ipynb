# Linear Regression example 1

```python
import pandas as pd
bmi_life_data = pd.read_csv('bmi_and_life_expectancy.csv')
bmi_life_data
```

```python
import matplotlib.pyplot as mp
mp.plot(bmi_life_data[['BMI']] , bmi_life_data[['Life expectancy']], 'bo')
mp.show()
```

```python
from sklearn.linear_model import LinearRegression
x = bmi_life_data[[bmi_life_data.columns[2]]]
y = bmi_life_data[[bmi_life_data.columns[1]]]
bmi_life_model = LinearRegression()
bmi_life_model.fit(x, y)
laos_life_exp = bmi_life_model.predict(60.31564)
laos_life_exp

```

```python
mp.scatter(x, y)
mp.plot(x, bmi_life_model.predict(x))
mp.show()
```
